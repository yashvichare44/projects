```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import category_encoders as ce

df = pd.DataFrame(data)

# Target Encoding for high cardinality column
target_encoder = ce.TargetEncoder(cols=['Car_Name'])
df['Car_Name_Encoded'] = target_encoder.fit_transform(df['Car_Name'], df['Selling_Price'])

# One-Hot Encoding for low cardinality column
df = pd.get_dummies(df, columns=['Fuel_Type'], prefix=['Fuel_Type'])
df['Fuel_Type_CNG'] = df['Fuel_Type_CNG'].astype(int)
df['Fuel_Type_Diesel'] = df['Fuel_Type_Diesel'].astype(int)
df['Fuel_Type_Petrol'] = df['Fuel_Type_Petrol'].astype(int)

# Label Encoding for binary columns
label_encoder = LabelEncoder()
df['Selling_type_Encoded'] = label_encoder.fit_transform(df['Selling_type'])
df['Transmission_Encoded'] = label_encoder.fit_transform(df['Transmission'])


# Drop original columns if not needed
df.drop(['Car_Name', 'Selling_type', 'Transmission'], axis=1, inplace=True)

# Print the resulting DataFrame
print(df)
```

```python
df.head()
```

```python

# Calculate the loss percentage
df['Loss_Percentage'] = ((df['Present_Price'] - df['Selling_Price']) / df['Present_Price']) * 100

# Calculate the cost price based on the selling price and loss percentage
df['Cost_Price'] = (100 / (100 - df['Loss_Percentage'])) * df['Selling_Price']

# Display the DataFrame with the calculated cost prices
print(df)
```

```python
df.drop(columns=['Loss_Percentage'], inplace=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

```

```python
X = df.drop(columns=['Selling_Price','Present_Price','Cost_Price','Owner'], axis=1)  # Features (all columns except Selling_Price)
y = df['Cost_Price']  # Target variable (Cost_Price)

# Split the data into training and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)

# Initialize the linear regression model
model = LinearRegression()

# Fit the model using the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
